```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

```python
xlsfile=pd.ExcelFile('daneEZR.xlsx')
```

```python
dane=xlsfile.parse(7)
```

```python
dane
```

```python
f, axs = plt.subplots(1,3,figsize=(20,5))
plt.subplot(131)
plt.plot(dane['Czas_lata'],dane['q'])
plt.title('Wykres q(t)')

plt.subplot(132)
plt.semilogy(dane['Czas_lata'],dane['q'])
plt.title('Wykres log(q(t))')

plt.subplot(133)
plt.loglog(dane['Czas_lata'],dane['q'])
plt.title('Wykres log(q(log(t)))')
```

```python
f, axs = plt.subplots(1,3,figsize=(20,5))
plt.subplot(131)
plt.plot(dane['Gp'],dane['q'])
plt.title('Wykres Gp(q)')

plt.subplot(132)
plt.semilogy(dane['Gp'],dane['q'])
plt.title('Wykres log(Gp(q))')

plt.subplot(133)
plt.loglog(dane['Gp'],dane['q'])
plt.title('Wykres log(Gp(log(q)))')
```

```python
qi=dane['q'][0]
q2=dane['q'][8]
q1=np.sqrt(qi*q2)
q1=np.array([q1,q1,q1,q1,q1,q1,q1,q1,q1])
plt.semilogy(dane['Czas_lata'],dane['q'])
plt.plot(dane['Czas_lata'],q1,marker='_')
plt.grid(True)
print('q1=',q1)
```
