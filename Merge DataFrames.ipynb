```python
import numpy as np 
import pandas as pd 
from pandas import Series, DataFrame
```

```python
dframe1=DataFrame({'key':['X','Y','Z','X','Z','Z'], 'data_set1':np.arange(6)})
dframe1
```

```python
dframe2=DataFrame({'key':['Q','Y','Z'],'data_set2':np.arange(3)})
dframe2
```

```python
pd.merge(dframe1,dframe2,on='key')
```

```python
dframe3=pd.merge(dframe1,dframe2,on='key',how='left')
dframe3
```
